<p>For the impatient:</p>

<p>A <em>function</em> in mathematics is <a href="http://tinyurl.com/c2rwqyb">defined</a> as "a relation between a set of inputs and a set of permissible outputs with the property that each input is related to exactly one output." That is a general definition. Specialized to mathematical functions of one real variable returning a real value, we can define a function in terms of a rule, such as:</p>


$$
f(x) = x^2 - 2.
$$


<p>The <strong>domain</strong> is the set of all permissible values for $x$, in this case all $x$, but this need not be the case either due to the rule not being defined for some $x$ or a more explicit restriction, such as $x \geq 0$. The <strong>range</strong> is the set of all possible outputs. Written in set notation, this is $\{f(x): x \in \text{ the domain }\}$.</p>

<p>Mathematically, we evaluate or call a function with the notation $f(2)$ or $f(3)$, say.</p>

<p>Mathematically we might refer to the function by its name, $f$, or its values $f(2)$, ...</p>

<p>In <code>Julia</code> basic mathematical  functions are defined and used with the <em>exact</em> same notation. This creates a function <code>f</code>:</p>

```julia
f(x) = x^2 - 2
```

<p>Unlike an expression, the value <code>x</code> in is not needed to be defined until we call the function. As with math, this variable name need not be <code>x</code> – it could be <code>y</code> or <code>theta</code> – though it often is.</p>

<p>We can call <code>f</code> for the value of <code>2</code> with:</p>

```julia
f(2)
```

<p>That is, as with typical mathematical notation, the function is "called" by passing a value to it with parentheses.</p>

<p>Within a cell, we can evaluate one or more values by using commas to separate them:</p>

```julia
f(1), f(2), f(3)
```

<p>The function name refers to the function object:</p>

```julia
f
```

<p>Don't worry about the words "generic" and "method", but be aware that because of this you can't rename a function into a variable, without an error. As well, <code>Julia</code> isn't even very keen on reusing a function name for another function and may give a warning.</p>

<p>Functions can be more complicated than the "one-liners" illustrated. In that case, a multiline form is available:</p>

```julia
function fn_name(args...)
  body
end
```

<p>The keyword <code>function</code> indicates this is a function whose name is given in the definition. Within the body, the last expression evaluated is the ouput, unless a <code>return</code> statement is used.</p>

<h2>Cases</h2>

<p>Some functions are defined in terms of cases. For example, a cell phone plan might depend on the data used through:</p>

<blockquote>
<p>The amount is 40 dollars for the first 1 Gb  of data, and 10 dollars more for each <em>additional</em> Gb of data.</p>
</blockquote>

<p>This function has two cases to consider: one if the data is less than 1 Gb and the other when it is more.</p>

<p>How to write this in <code>julia</code>?</p>

<p>The ternary operator <code>predicate ? expression1 : expression2</code> has three pieces: a predicate question, such as <code>x &lt; 10</code> and two expressions, the first is evaluated if the predicate is <code>true</code> and the second if the predicate is <code>false</code>. They are useful to write functions that are defined by cases. (They are a light-weight form of the traditional <code>if-then-else</code> construct.)</p>

<p>The example above, could then be done with:</p>

```julia
f(d) = d <= 1.0 ? 40.00 : 40.00 + 10.00 * (d - 1.0)
```

<h2>Parameters: Functions can have keyword arguments.</h2>

<p>In the following, we define a function which evaluates a linear equation. A familiar description of a line has two parameters: the slope $m$ and the $y$ intercept, $b$: $y=mx + b$. The main focus is on $x$, but the values of $m$ and $b$ give each problem a different application. To keep these separate, keyword arguments may be given. In the following, <code>m</code> and <code>b</code> are passed in via keywords and, as defined, have default values of <code>m&#61;1</code> and <code>b&#61;0</code>.</p>

```julia
mxplusb(x; m=1, b=0) = m*x + b
mxplusb(10, m=3, b=4)		# arguments are named
```

<p>The semicolon separates the two types of arguments.</p>

<p>If no value are passed, the defaults are used:</p>

```julia
mxplusb(10, m=2)   # evaluates 2 * 10 + 0
```

<h2>Functions can be used as arguments to other functions:</h2>

<p>Like many computer languages, <code>Julia</code> allows functions to be arguments to functions. In the following, we see we can plot a function by passing in a function object to a <code>plot</code> function. Notice, the variable name (<code>f</code>) and not the call (<code>f&#40;x&#41;</code>) is given to the <code>plot</code> command (from the <code>Plots</code> package, loaded with <code>MTH229</code>):</p>

<h4>Functions can be anonymous.</h4>

<p>When functions are passed into other functions, it can be convenient to define them without names. The syntax is slightly different, but follows mathematical notation: <code>x -&gt; expression_in_x</code>, or more generally <code>args -&gt; body</code>. Such functions are called <em>anonymous functions</em>:</p>

```julia
plot(x -> sin(x)*cos(2x), 0, 2pi)
```

<h2>Some other facts about functions</h2>

<h4>Julia functions can have more than one variable</h4>

<p>For example, this function is used to compute the area of a rectangle:</p>

```julia
rectarea(b, h) = b * h		# area of rectangle is base times height
```

<p>The <code>log</code> function is an example, where <code>log&#40;b,x&#41;</code> will find the log base $b$ of $x$, while <code>log&#40;x&#41;</code> uses the default base <code>e</code>, or the natural log.</p>

<h4>Julia functions with different signatures, can have the same name!</h4>

<p>This finds the area of a square using the previously defined rectangle function:</p>

```julia
rectarea(b) = rectarea(b,b)	# area of square using area of rectangle
```

<p>Which function is used depends on the arguments that are passed when calling the function. Hence, <code>log&#40;x&#41;</code> will find log base $e$ of $x$ whereas <code>log&#40;10, x&#41;</code> can be used to find log base 10.</p>
